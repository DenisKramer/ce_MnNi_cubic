# Convex Hull

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull, convex_hull_plot_2d

df=pd.read_fwf('../fit_1/summary.dat',header=0)

points = np.array([[v[0],v[1]] for v in zip(df['comp(a)'],df['formation_energy'])])
hull = ConvexHull(points)

plt.figure(figsize=(12,9))
plt.plot(df['comp(a)'],df['formation_energy'],'x')
for simplex in hull.simplices:

    if not np.all(points[simplex, 1] == 0.0):
        plt.plot(points[simplex, 0], points[simplex, 1], 'ro-')
        
plt
```

# Correlation plot

## Full concentration range
This is a cluster expansion over the full concentration range. Given the complexity of the charge transfer in the system, the quality of the fit is not that great.

```python
df=pd.read_fwf('../fit_1/summary.dat',header=0)
plt.figure(figsize=(7,7))
plt.plot(df['formation_energy'],df['clex()'],'x')
plt.plot([min(df['formation_energy']),max(df['formation_energy'])],[min(df['formation_energy']),max(df['formation_energy'])],'r-')
plt.xlabel("formation energy DFT [eV]");
plt.ylabel("formation energy CE [eV]");
```
